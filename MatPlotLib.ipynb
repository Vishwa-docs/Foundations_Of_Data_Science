### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graphs
##### [Plot Function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)
##### [Title](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html#matplotlib.pyplot.title)
##### [xlabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html#matplotlib.pyplot.xlabel) and [ylabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html#matplotlib.pyplot.ylabel)

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, label = '2x', c = 'red', linewidth = 2, marker = '.', linestyle = '--', markersize = 10, markeredgecolor = 'blue') # x and y as 1D arrays, color can be Hex Hash

'''
You can use the shorthand notation
'[color][marker][line]'

eg : 
plt.plot(x, y, 'c.--', label = '2x')
'''

plt.title('Graph', fontdict = {'fontname' : 'Comic Sans MS', 'fontsize' : 15})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# To add the scale labels. The size between ticks will be resized according to the numbers given as input
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([1, 2, 3, 4, 8])


plt.legend() # Only works with the label handle in plt.plot()

plt.show() # Not Required
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

# Resizing the Graph : 
plt.figure(figsize=(5, 3), dpi=300) # dpi is pixels per inch


plt.plot(x, y, 'c.--', label = '2x')

# Adding the Second line to above graph : 
x2 = np.arange(0, 4, 0.5)
print(x2)
#plt.plot(x2, x2**2, 'r', label = 'x^2')

# Dividing the line into 2 graphs
plt.plot(x2[:5], x2[:5]**2, 'r', label = 'x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

# Save the Graph
plt.savefig('mygraph.png', dpi = 300) # If you dont want the dpi to increase originally, you can pass it as a parameter here

plt.show()
```

### Bar Charts

```python
labels = ['A', 'B', 'C']
vlaues = [1, 4, 2]

bars = plt.bar(labels, vlaues)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

'''
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')
'''

plt.figure(figsize = (6, 4))

plt.show()
```

### Line Plot

```python
# Read the file
data = pd.read_csv('name.csv')
print(data) 
# or print(data.head(5)) will print only first 5 rows

plt.title('Name')

# Both lines will be plotted on same graph
plt.plot(data.Year, data.Country)
plt.plot(data['Second Year'], data.Country) # For 2 words

# OR : 
for country in data:
    plt.plot(data.Year, data.Country, marker = '.') # Different colours will come
# You can choose specific countries you want as a list and then iterate through them

plt.xticks(data.Year[::3].tolist() + [2011]) # Ever 3 years in x axis from beginning to end, with 2011 as the last year

plt.legend() # Dont need to add label with CSV Files
# You can move legend outside (Google it)

plt.show()
```

### Histogram

```python
#matplotlib.pyplot.hist

bins = [0, 10, 20, 30]

plt.hist(data.Year, bins = bins, color = '#abcdef') # You can google color picker to find the hex colour

plt.xticks(bins)

plt.show()
```

### Pie Charts

```python
left = data.loc[data['Preferred'] == 'Left'].count()[0]
right = data.loc[data['Preferred'] == 'Right'].count()[0]

labels = ['left', 'right']
colors = ['#5thf7r', '#abcdef']
explode = (.4, .6)
plt.pie([left, right], labels = labels, colors = colors, autopct = '%.2f %%', pctdistance = 2.0, explode = explode) # pctdistance is distance of the percentage from centre, explode will split the graphs

data.Weight = [int(x.strip('lbs')) if type(x) == str ele x for x in data.Weight]
plt.style.use('ggplot') # A preset colour theme for pie charts
plt.show()
```
